```python
def sigmoid(z):
    return 1/(1+np.exp((-1)*z))
```

```python
def hy(x,weights):
    z=np.dot(x,weights)
    #print(z.shape)
    return sigmoid(z)
    
    
```

```python
def log_likelihood(X,Y,weights,e=0.0000001):
    m=len(Y)
    
    hypothesis=hy(X,weights)
    print(" hypothesis",(hypothesis))
    error=1*Y*np.log(hypothesis+e)+(1-Y)*np.log(1-hypothesis+e)
    #print(error)
    
    cost=-np.mean(error)
    return cost
```

```python
def gradient(X,Y,weights):
    hypothesis=hy(X,weights)
    derivative=np.dot(X.T,(hypothesis-Y))
    return derivative/len(X)
```

```python
def predict(x,weights):
    m=len(x)
    pred=np.zeros((m,1))
    temp=hy(x,weights)
    pred = np.where(temp > 0.5, 1, 0)    
    return pred
```

```python
def gradient_descent(X,Y,weights,learning_rate=0.001,num_iterations=100):
    
    cost_history=[]
    grad_history=[]
    weights_history=[]
    #print(1)
    for i in range(num_iterations):
        hypothesis=hy(X,weights)
        #print(2)
        cost=log_likelihood(X,Y,weights)
        #print(3)
        cost_history.append(cost)
        #print(4)
        grad=gradient(X,Y,weights)
        grad_history.append(grad)
        #print(5)
        #print("grad",grad)
        weights=weights-learning_rate*grad
        #print(6)
        #print("weights",weights)
        weights_history.append(weights)
        #print(7)
        
        print("cost itr",i, cost)
        if cost_history[i-1]-cost_history[i]<10**-7 and cost_history[i-1]-cost_history[i]>0:
            plt.plot(list(range(i)),cost_history)
         
            return weights
    plt.plot(list(range(i+1)),cost_history)
    plt.title('cost')
    return weights    
```

```python
n=train_x.shape[1]

train_x.insert(0,'intercept',np.ones(len(train_x)),True)
```

```python
weights=np.zeros((n+1,1))

train_y=np.array(train_y).reshape(len(train_y),1)
weights=gradient_descent(train_x,train_y,weights,learning_rate=0.001,num_iterations=2200)
```
